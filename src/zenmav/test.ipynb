```python

from core import Zenmav
import time
drone = Zenmav(gps_thresh=0.5)
drone.guided_arm_takeoff(25)

```

```python
drone.set_mode('ACRO')
finished = False
start_time = time.time()
while not finished :
    if time.time()- start_time < 1.0:
        drone.rc_override({'ch3': 1850})
        print('OVERRIDE')
    elif  3.14 > drone.get_attitude()[0] > -0.1:
        drone.rc_override({'ch3': 1000, 'ch1': 2000})
    elif -3.14 < drone.get_attitude()[0]< -0.2:
        drone.rc_override({'ch3': 1000, 'ch1': 1800})
    else:
        drone.set_mode('GUIDED')
        finished = True 
        print('FINISHED')
```
